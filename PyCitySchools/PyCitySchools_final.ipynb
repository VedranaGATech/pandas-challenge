```python
#file paths 
schools_data_file = "Resources/schools_complete.csv"
students_data_file = "Resources/students_complete.csv"
```

```python
#read files and store into panda dataframe 
schools_dataDF = pd.read_csv(schools_data_file)
schools_dataDF.head(3)
```

# District Summary

Create a high-level snapshot of the folowing district's key metrics in a DataFrame:
Total schools, Total students, Total budget, Average math score, Average reading score, % passing math, % passing reading, and %overall passing

```python
#Total number of schools
total_number_schools = len(schools_dataDF["School ID"].unique())
total_number_schools
```

```python
#Total number of students
total_number_students = len(students_dataDF["Student ID"].unique())
total_number_students
```

```python
#Total school budget
total_budget = schools_dataDF["budget"].sum()
total_budget
```

```python
#Average math score
avg_math_score = students_dataDF["math_score"].mean()
avg_math_score
```

```python
#Average reading score
avg_reading_score = students_dataDF["reading_score"].mean()
avg_reading_score
```

```python
#count of students passing math
students_passing_math= students_dataDF.loc[students_dataDF["math_score"] >= 70, "math_score"]
students_passing_math.count()
```

```python
#% of students passing math - (count of total students passing math/total students)*100
percent_passing_math = float(students_passing_math.count()/ total_number_students)*100
percent_passing_math
```

```python
#count of students passing reading
students_passing_reading= students_dataDF.loc[students_dataDF["reading_score"] >= 70,"reading_score"]
students_passing_reading.count()
```

```python
#% of students passing reading -(count of total students passing readin/total students)*100
percent_passing_reading = float(students_passing_reading.count()/total_number_students)*100
percent_passing_reading
```

```python
#% overall passing - the percentage of students who passed math and reading
#first get the count of overall number of students passing math AND reading 
overall_passing_math_reading = students_dataDF[(students_dataDF["math_score"] >= 70) & (students_dataDF["reading_score"] >= 70)]
overall_passing_math_reading["student_name"].count()
overall_passing_count = overall_passing_math_reading["student_name"].count()
percent_overall_passing = (overall_passing_count/total_number_students)*100
percent_overall_passing
```

```python
#Create a high-level snapshot of the district's key metrics in a DataFrame
    #Total schools
    #Total students
    #Total budget
    #Average math score
    #Average reading score
    #% passing math
    #% passing reading
    #% overall passing
    
district_summaryDF = pd.DataFrame(
    {"Total Schools":[total_number_schools],
     "Total Students":[total_number_students],
     "Total Budget":[total_budget],
     "Average Math Score":[avg_math_score],
     "Average Reading Score":[avg_reading_score],
     "% Passing Math":[percent_passing_math],
     "% Passing Reading":[percent_passing_reading],
     "% Overall Passing":[percent_overall_passing]
     })


# Added  formating 
district_summaryDF["Total Students"] = district_summaryDF["Total Students"].map("{:,}".format)
district_summaryDF["Total Budget"] = district_summaryDF["Total Budget"].map("${:,.2f}".format)
district_summaryDF["Average Math Score"] = district_summaryDF["Average Math Score"].map("{:.2f}".format)
district_summaryDF["Average Reading Score"] = district_summaryDF["Average Reading Score"].map("{:.2f}".format)
district_summaryDF["% Passing Math"] = district_summaryDF["% Passing Math"].map("{:.4f}".format)
district_summaryDF["% Passing Reading"] = district_summaryDF["% Passing Reading"].map("{:.4f}".format)
district_summaryDF["% Overall Passing"] = district_summaryDF["% Overall Passing"].map("{:.4f}".format)

district_summaryDF
```

# School Summary

Create a DataFrame that summarizes the following key metrics about each school: School name, School type, Total students, Total school budget, Per student budget, Average math score, Average reading score, % passing math, % passing reading, and % overall passing.


```python
#test print
#schools_data_complete.head(5)
```

```python
#get school types 
school_type = schools_data_complete.groupby("school_name")["type"].unique()
#school_type
```

```python
#get per school student count 
total_students = schools_data_complete["school_name"].value_counts()
#total_students
```

```python
#get per school budget
school_budget = schools_data_complete.groupby("school_name")["budget"].mean()
#school_budget
```

```python
#get per school per student budget
per_student_budget = school_budget/total_students
#per_student_budget
```

```python
#per school average math score
avg_math_score = schools_data_complete.groupby("school_name")["math_score"].mean()
#avg_math_score
```

```python
#per school average reading score
avg_reading_score = schools_data_complete.groupby("school_name")["reading_score"].mean()
#avg_reading_score
```

```python
#per school % passing math
passing_math = schools_data_complete[schools_data_complete["math_score"] >= 70].groupby("school_name")["Student ID"].count()
prc_passing_math = (passing_math/total_students)*100
#prc_passing_math
```

```python
#per school % passing reading
passing_reading = schools_data_complete[schools_data_complete["reading_score"] >= 70].groupby("school_name")["Student ID"].count()
prc_passing_reading = (passing_reading/total_students)*100
#prc_passing_reading
```

```python
#get the count - per school passing both math and reading per school 
per_passing_math_reading = schools_data_complete[(schools_data_complete["math_score"] >= 70) & (schools_data_complete["reading_score"] >= 70)]
per_passing_math_reading = per_passing_math_reading.groupby(["school_name"]).count()["student_name"]
#per_passing_math_reading
```

```python
#per school % overall passing 
prc_overall_passing = (per_passing_math_reading / total_students) * 100
#prc_overall_passing
```

```python
#per School Summary DataFrame

schools_summaryDF = pd.DataFrame (
        {"School Type" : school_type,
         "Total Students" : total_students,
         "Total School Budget" : school_budget,
         "Per Student Budget" : per_student_budget, 
         "Average Math Score" : round(avg_math_score,2),
         "Average Reading Score" : round(avg_reading_score,2),
         "% Passing Math" : round(prc_passing_math,2),
         "% Passing Reading" : round(prc_passing_reading,2),
         "% Overall Passing" : round(prc_overall_passing,2),
            
        }
)

#Formatting
schools_summaryDF["Total Students"] = schools_summaryDF["Total Students"].map("{:,.0f}".format)
schools_summaryDF["Total School Budget"] = schools_summaryDF["Total School Budget"].map("${:,.2f}".format)
schools_summaryDF["Per Student Budget"] = schools_summaryDF["Per Student Budget"].map("${:,.2f}".format)

schools_summaryDF
```
